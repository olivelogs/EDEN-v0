# Get Bounds from Geopackage (generated with prep_ecoregions.py)

### Get bounds of all regions


```python
# Load your geopackage
gdf = gpd.read_file('/Users/olivelo/Desktop/EDEN-v0/data/interim/vectors/ecoregions_selected.gpkg')

# For each region, get bounds
for idx, row in gdf.iterrows():
    bounds = row.geometry.bounds  # returns (minx, miny, maxx, maxy)
    print(f"{row['name']}: {bounds}")
```

### Get bounds -> unify regions

```python
gdf = gpd.read_file('/Users/olivelo/Desktop/EDEN-v0/data/interim/vectors/ecoregions_selected.gpkg')

# Group by region name and dissolve into single geometry
for name in gdf['name'].unique():
    region_subset = gdf[gdf['name'] == name]
    
    # Dissolve all geometries into ONE unified shape
    unified = region_subset.dissolve()
    
    # NOW get bounds (will be one bounding box for whole region)
    bounds = unified.geometry.bounds.iloc[0]
    print(f"{name}: {list(bounds)}")
```
